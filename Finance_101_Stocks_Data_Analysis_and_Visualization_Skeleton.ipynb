- **Print out normalized (scaled) stock prices**  


```
def normalized(df):  #the func passes a dataframe as parametre
  x = df.copy()  #making copy of the dataframe to manipulate it without causing change to the actual dataframe
  for i in x.columns[1:]:  #we do not need the date column hence "columns [1:]"
    x[i] = x[i]/x[i][0] #divide value in each index of the column with the first value of the columns
  return x
```

# TASK #4: PERFORM INTERACTIVE DATA VISUALIZATION

```
# Function to perform an interactive data plotting using plotly express
# Plotly.express module which is imported as px includes functions that can plot interactive plots easily and effectively. 
# Every Plotly Express function uses graph objects internally and returns a plotly.graph_objects.Figure instance. 

def interactive_plot (df, title):

  fig = px.line(title = title) #using plotly

  for i in df.columns[1:]:
    fig.add_scatter(x = df['Date'], y = df [i] , name = i)

  fig.show()  
```

# TASK #5: CALCULATE INDIVIDUAL STOCKS DAILY RETURNS

```
# Let's calculate daily return for a single security
# Let's take the S&P500 as an example first
sp500_df = stocks_df['sp500']
sp500_df
```

```
df_daily_return = sp500_df.copy()

df_daily_return [0] = 0

for j in range(1, len(sp500_df)):
  df_daily_return[j] = ((sp500_df[j] - sp500_df[j-1]) /sp500_df[j])*100                 #calc daily return in % current price minus yesterdays price divided by current price percent. 

df_daily_return
```

# TASK #6: CALCULATE MULTIPLE STOCKS DAILY RETURNS

```
# Let's define a function to calculate stocks daily returns (for all stocks) 
def stocks_daily_return (df):
  df_daily_return = df.copy()
  
  #loop for clmns

  for i in df.columns[1:]:

    df_daily_return [i][0] = 0 
  
    #loop for rows
    for j in range(1, len(df)):
      df_daily_return[i][j] = ((df[i][j] - df[i][j-1]) /df[i][j-1])*100

    
  return df_daily_return
```

```
# Get the daily returns 
stocks_daily_return = stocks_daily_return(stocks_df)
```
